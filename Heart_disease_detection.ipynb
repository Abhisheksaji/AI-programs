```python
# replacing ? and replacing it with mean
df['age'] = df['age'].replace(to_replace='?', value=df['age'][df['age'] != '?'].astype(float).mean())
df['sex'] = df['sex'].replace(to_replace='?', value=df['sex'][df['sex'] != '?'].astype(float).mean())
df['cp'] = df['cp'].replace(to_replace='?', value=df['cp'][df['cp'] != '?'].astype(float).mean())
df['trestbps'] = df['trestbps'].replace(to_replace='?', value=df['trestbps'][df['trestbps'] != '?'].astype(float).mean())
df['chol'] = df['chol'].replace(to_replace='?', value=df['chol'][df['chol'] != '?'].astype(float).mean())
df['fbs'] = df['fbs'].replace(to_replace='?', value=df['fbs'][df['fbs'] != '?'].astype(float).mean())
df['restecg'] = df['restecg'].replace(to_replace='?', value=df['restecg'][df['restecg'] != '?'].astype(float).mean())
df['thalach'] = df['thalach'].replace(to_replace='?', value=df['thalach'][df['thalach'] != '?'].astype(float).mean())
df['exang'] = df['exang'].replace(to_replace='?', value=df['exang'][df['exang'] != '?'].astype(float).mean())
df['oldpeak'] = df['oldpeak'].replace(to_replace='?', value=df['oldpeak'][df['oldpeak'] != '?'].astype(float).mean())
```

```python
#df[['trestbps','chol','fbs','restecg','thalach','exang']]
#converting into int
df['trestbps']=df['trestbps'].astype(int)
df['chol']=df['chol'].astype(int)
df['fbs']=df['fbs'].astype(int)
df['restecg']=df['restecg'].astype(int)
df['thalach']=df['thalach'].astype(int)
df['exang']=df['exang'].astype(int)
```

```python
df.info()
```

```python
x=df.drop(columns=['num','slope','ca','thal'],axis=1)
y=df['num']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_train)
```

```python
score=accuracy_score(prediction,y_train)
```
